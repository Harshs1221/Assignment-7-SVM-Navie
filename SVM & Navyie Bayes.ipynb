```python
#Q1. What is a Support Vector Machine (SVM)?
"""
Ans1. A Support Vector Machine is a powerful supervised learning algorithm used for classification and
      regression.
      It works by finding the optimal hyperlane that best separates different classes in a dataset.
"""
```

```python
#Q2. What is the difference between Hard Margin and Soft Margin SVM?
"""
Ans2. The difference between hard margin and soft margin are:
      1. Hard Margin SVM
          - Used when the data is perfectly separable.
          - It finds the maximum hyperplane without any misclassification.
          - Strict rule: Every data point must be correctly classified.
          - Limitation:
                If there's even a little noise or overlap between classes, Hard Margin SVM fails.
      2. Soft Margin SVM
          - Used when data is not perfectly separable.
          - Allows some misclassification to improve generalization.
          - More flexible: Works better with the real-world noisy data.
"""
```

```python
#Q3. What is the mathematical intuition behind SVM?
"""
Ans3. The mathematical intution behind Support Vector Machines (SVM) revolves around fining the optimal 
      hyperplane that best separates different classes while maximizing the margin between them.

      1. Hyperplane Equation:
                  w.x + b = 0
            where w is the wight vector, x in the input vector and b is bias.
      2. Margin Maximization:
            SVM aims to maximize the margin between the hyperplane and closest data points. The margin is
            given by:
                                    2 / ||w||
            A larger margin leads to better generalization.
      
      3. Optimization Problem:
            SVM solves the following optimization problem:
                        min ùë§,ùëè 1/2‚à£‚à£ùë§‚à£‚à£**2
            subject to the constraint:
                        yi(w.xi + b) >= 1                        
"""
```

```python
#Q4.  What is the role of Lagrange Multipliers in SVM?
"""
Ans4. Lagrange helps SVM find the best dividing line between classes while hadnling constraints.
        Like:
            - SVM wants to maximize the margin between different classes.
            
            - But it also needs to respect the constraints (data must be correctly modified)
            
            - Lagrange Multipliers ast as a balancing tool- they help SVM find the best solution
              while keeping the constraints in check.
"""
```

```python
#Q5. What are Support Vectors in SVM?
"""
Ans5. Support vectors are the key data points in a Support Vector Machine (SVM) that help define the 
      optimal hyperplane for classification.
      These are the closest points to the decision boundary and have a direct impact on its position.
"""
```

```python
#Q6. What is a Support Vector Classifier (SVC)?
"""
Ans6. A Support Vector Classifier(SVC) is a type of Support vector Machine used for classificatio tasks. It finds
      the optimal hyperplane that separates different classes while maximizing the margin between them.
      - It works well with binaru and multi-class classification.
      - Uses support vector to define the decision boundary.
      - Can handle linear and non-linear classification using different kernels.
      - Uses regularization( C parameter) to control the trade-off between margin and misclassification.
"""
```

```python
#Q7. What is a Support Vector Regressor (SVR)?
"""
Ans7. Support Vector Regressor is a type of Support Vector Machine used for regression tasks. It predict the 
      continous values by finding a hyperplane that best fits the data while minimizing errors.
      
      - Works well for linear and non-linear regression using different kernels (e.g., linear, polynomial,RBF)
      - Uses support vectors to define the best-fit function.
      - Introduces an epsilon-tube, meaning it ignores small errors within a certain range to improve 
        generalization.
      - uses regularization (C parameter) to control the trade-off between complexity and accuracy.
"""
```

```python
#Q8.  What is the Kernel Trick in SVM?
"""
Ans8. The kernel trick in SVM helps solve problems where data is not linearly separable. 
      Instead of manually transforming the data into higher dimensional space, kernel function does this
      automatically, allowing SVM to find the better decision boundary.

      Types of Kernels:
          1. Linear - Better for simple, linearly separable data.
          2. Polynomial Kernel - Useful when data has curved patterns
          3. RBF(Radial Based Function) Kernel - Works well for complex datasets
"""
```

```python
#Q9.  Compare Linear Kernel, Polynomial Kernel, and RBF Kernel.
"""
Ans9. Support vector machines use kernels to transform data into higher-dimensional spaces for better 
      classification.
      - Linear Kernel is best for simple, linearly separabel data.
      - Polynomial Kernal is useful when the decision boundary is corved.
      - RBF Kernel is most flexible, handling complex, non-linear data effectively.
"""
```

```python
#Q10. What is the effect of the C parameter in SVM?
"""
Ans10. The C parameter in SVM controls the trade-off between maximizing the margin and minimizing 
       classification errors.

       Effect of C Parameter:

       1. Small C Value (Large Margin, More Misclassifications)
           - Allows more misclassified points in training.
           - Focuses on maximizing the margin rather than fitting every data point perfectly.
           - Leads to better generalization but may underfir if too small.
       
       2. Large C Value (Narrow Margin, Fewer Misclassification)
           - Tries to classify all training points correctly.
           - Leads to narrower margin, making the model more sensitive to noise.
           - Can overfit if too large, meaning it performs well on training data but poorly on new data.
"""
```

```python
#Q11. What is the role of the Gamma parameter in RBF Kernel SVM?
"""
Ans11. The Gamma Parameter (y) in the RBF Kernel of SVM controls how much influence each data point has
       on the decision boundary.
       - Low Gamma (Small y) -
               Each point has a wide influence, leading to a smooth boundary. This helps in better 
               generalization but may underfit the data.
       - high Gamma (Large y) - 
               Each point has a narrow influence, making the boundary more complex. This can overfit the data,
               meaning it works well on training data but poorly on new data.
"""
```

```python
#Q16.  What are the key assumptions made by Na√Øve Bayes?
"""
Ans16. Naive Bayes makes two key assumption that simplify probability calculation for classifiation:

        1. Feature Independence Assumption:
            - It assumes that all features are independent given the class label.
            - Example: In spam detection, Naive bayes assumes that workd like "free" and "offer"
                       appear independently, even though they often occur together in spam emails.
              Reality: Features are often correlated, but this assumption , make computation efficient.
        
        2. Equal Contribution Assumption
            - It assumes that each feature contributes equally to the final classification.
            - Example: In a medical diagnosis model, Naive Bayes treats age and blood pressure
                       as equally important, even though some features may have more impact.
            - Reality: Some features are influential, But Naive bayes does not weigh them
                       differently.
"""
```

```python
#Q20.  How does Laplace Smoothing help in Na√Øve Bayes?
"""
Ans20. Leplace smoothing helps Naive Bayes by preventing zero probabilities when a word or feature does
       not appear in the training data.
       Without smoothing, if a word is missing from a class, its probability becomes zero, which can make 
       the entire classification fail.

       - it adds a small constant(usually 1) to all word counts.
       - This ensure that no probability is ever zero.

       Example:
           - If the word "discount" never appeared in spam emails during training, its probability 
             would be zero.
           - Without smoothing, any new email containing "discount" would never be classified as spam.
           - Laplace Smoothing ensures that even unseen words get a small probability, allowing better 
             prediction.
"""
```

```python
#Q27. Write a Python program to train an SVM Classifier with different C values and compare the decision boundaries visually
```

```python
X, y = make_classification(n_samples=200, n_features=2, n_classes=2,n_redundant=0,n_informative=2, n_clusters_per_class=1, random_state=1)
```

```python
#train test split
X_train,X_test, y_train,  y_test =  train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
C_value = [0.1, 1, 10]
```

```python
fig, axes = plt.subplots(1, len(C_value), figsize=(15,6))
```

```python
# Model Training
for i, C in enumerate(C_value):
    model = SVC(kernel='linear', C=C)
    model.fit(X_train, y_train)
    #compute decision boundary
    w = model.coef_[0]
    a = -w[0] / w[1]
    xx = np.linspace(X[:, 0].min() -1, X[:,0].max() +1, 100)
    yy = a * xx - (model.intercept_[0] / w[1])

    sns.scatterplot(x= X_test[:,0], y= X_test[:,1], hue = y_test, ax = axes[i])
    axes[i].plot(xx,yy, color = 'red')
    axes[i].set_title(f"SVM Decision boundary (C = {C})")
plt.tight_layout()
plt.show()
```
